# Prise en compte de l'écrouissage

Sans écrouissage, on observe que la plasticité se localise sur le bord interne. Cela se voit analytiquement par l'impossibilité d'écrire à la fois l'admissibilité statique ($\sigma \approx \dfrac{1}{r}$) et le critère de plasticité ($\sigma = constante$).

L'idée est donc de prendre en compte de l'écrouissage, ce qui permettrait de "débloquer" un degré de liberté. On choisit pour commencer un écrouissage isotrope linéaire durcissant $R(p) = R.p$ avec $R \in \mathbb{R}^+$.

On observe que plus R est grand, plus le front de plastification s'éloigne du bord interne. Ce qui est logique car pour R grand, la zone plastique résiste plus, et donc transmet plus d'efforts à la zone élastique qui plastifie plus vite, d'où la propagation plus "rapide". Ce qui permet également de mieux comprendre le cas $R = 0$ : dès que la couronne interne atteint la plasticité, cette zone permet un écoulement plastique infini, il n'y a donc plus rien qui est transmis au reste du matériau.

```python
from solvers_GB import solve_antiplane
import numpy as np
import matplotlib.pyplot as plt

Rs = np.linspace(0,0.6,6)



for i in range(len(Rs)):
    loads, Fi, Fe, plastic_dissipations = solve_antiplane(
                                                    R_i = 0.5,
                                                    R_e = 1.,
                                                    sig_0 = 1.,
                                                    R = Rs[i],
                                                    lc =.05,
                                                    dist_min = .1,
                                                    dist_max = .3,
                                                    refinement_ratio = 1.,
                                                    results = True,
                                                    results_name = "none",
                                                    c_lim = 0.05, # pour bien voir le front de plastification
                                                    t_max = 1.,
                                                    incr = 15
                                                    )
    print("R = ")
    print(Rs[i])
```

# Convergence du maillage



```python
from solvers_GB import solve_antiplane
import numpy as np
import matplotlib.pyplot as plt

rs = [1, 2, 3, 4]
p_last = []

for i, r in enumerate(rs):
    loads, Fi, Fe, plastic_dissipations = solve_antiplane(
                                                        R_i = 0.5,
                                                        R_e = 1.,
                                                        sig_0 = 1.,
                                                        R = 0.2,
                                                        lc =.2,
                                                        dist_min = .1,
                                                        dist_max = .3,
                                                        refinement_ratio = r,
                                                        results = True,
                                                        results_name = "none",
                                                        c_lim = 0.05,
                                                        t_max = 1.,
                                                        incr = 15
                                                        )
    p_last.append(plastic_dissipations[-1])

plt.plot(rs, p_last, 'o--')
plt.xlabel("Refinement ratio")
plt.ylabel("Final p")
plt.title("Convergence of plastic dissipation while refining the mesh")
plt.show()
```
